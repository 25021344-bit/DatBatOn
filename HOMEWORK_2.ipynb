#### Bài 3
Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là 
độ dài ba cạnh hay không, nếu  không thì in  ra màn  hình  'Khong la ba  canh 
cua mot tam  giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, 
Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn 
hình.

```python
a = float(input())
b = float(input())
c = float(input())
if a + b>c and c+b>a and a+c>b:
    if a==b==c:
        print("Đây là tam giác đều!")
    elif a==b or b==c or a==c:
        print("Đây là tam giác cân!")
    elif a**2+b**2==c**2 or a**2+c**2==b**2 or b**2+c**2==a**2:
        if a==b or a==c or b==c:
            print("Đây là tam giác vuông cân")
        else:    
            print("Đây là tam giác vuông!")
    else:
        print("Đây là tam giác thường")
else:
    print("Đây không là tam giác!")
```

#### Bài 5
 Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
a = float(input())
b = float(input())
c = float(input())
print(f'=> Chúng ta có phương trình bậc 2 là {a}x^2+{b}x+{c}=0')
print("+> Giai phuong trinh:")
if b%2 ==0:
    delta = ((b/2)**2) - a*c
    if delta > 0:
        print(f'Phuong trinh co hai nghiem phan biet la: x1 = {(-b/2 + delta**(1/2))/a} va x2 = {(-b/2 - delta**(1/2))/a}')
    elif delta == 0:
        print(f'Phuong trinh co hai nghiem phan biet la: x1 = x2 = {(-b/2 )/a}')
    else:
        print("Phuong trinh nay vo nghiem")
else:
    delta_2 = b**2 - 4*a*c
    if delta_2 > 0:
        print(f'Phuong trinh co hai nghiem phan biet la: x1 = {(-b + delta_2**(1/2))/2*a} va x2 = {(-b - delta_2**(1/2))/2*a}')
    elif delta_2 == 0:
        print(f'Phuong trinh co hai nghiem phan biet la: x1 = x2 = {(-b)/2*a}')
    else:
        print("Phuong trinh nay vo nghiem")
```
